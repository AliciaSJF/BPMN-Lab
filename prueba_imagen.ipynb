```python
import json
import pandas as pd
import xml.etree.ElementTree as ET
import io
from IPython.display import Image, display
from sapsam.RepresentationGenerator import RepresentationGenerator
from sapsam.SignavioAuthenticator import SignavioAuthenticator # üîπ Asegurar import correcto
from sapsam.SignavioAuthenticator import SignavioAuthenticator


with open('c:/Users/AliciaSanJuli√°nFerna/Documents/proyectos/process-modeling/modelo.json', 'r') as file:
    model_json = json.load(file)
model_namespace = "http://b3mn.org/stencilset/bpmn2.0#"
model_name = "fun"
print('================ Namespace:', model_namespace)
print('================ Name:', model_name)


gen = RepresentationGenerator()

# # üîπ Generar imagen del modelo

image_request = gen.generate_image(model_name, model_json, model_namespace, deletes=False)
display(Image(image_request, height=50, width=50, format='png'))

# üîπ Generar BPMN 2.0 XML
xml_request = io.StringIO(gen.generate_bpmn_xml(model_name, model_json, model_namespace,deletes=False).decode("utf-8"))
xml_tree = ET.parse(xml_request)
root = xml_tree.getroot()

# üîπ Mostrar el XML generado
print("\n",ET.tostring(root, encoding="unicode", method="xml"))
```
